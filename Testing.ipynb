```python
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.models import load_model
model = load_model("pcos_detection_model.h5")
```

```python
def preprocess_single_image(img_path, img_size=(128, 128)):
    img = cv2.imread(img_path)
    img = cv2.resize(img, img_size)
    img = img / 255.0 
    img = np.expand_dims(img, axis=0)  
    return img
```

```python
test_img_path = "new.png" 
img=Image.open(test_img_path)
plt.imshow(img)
plt.axis(False)
plt.title("UPLOADED IMAGE")
plt.show()
test_img = preprocess_single_image(test_img_path)
prediction = model.predict(test_img)
class_label = np.argmax(prediction, axis=1)
if class_label == 0:
    print("Prediction: Infected")
else:
    print("Prediction: Non-Infected")
```
